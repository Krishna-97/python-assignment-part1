# **Theme: Student Grade Tracker**
**Task 1 — Data Parsing & Profile Cleaning**


```python
raw_students = [
    {"name": "  ayesha SHARMA  ", "roll": "101", "marks_str": "88, 72, 95, 60, 78"},
    {"name": "ROHIT verma",       "roll": "102", "marks_str": "55, 68, 49, 72, 61"},
    {"name": "  Priya Nair  ",    "roll": "103", "marks_str": "91, 85, 88, 94, 79"},
    {"name": "karan MEHTA",       "roll": "104", "marks_str": "40, 55, 38, 62, 50"},
    {"name": " Sneha pillai ",    "roll": "105", "marks_str": "75, 80, 70, 68, 85"},
]

cleaned_students =[]
for s in raw_students:
  #cleaned_name has names with removed whitespaces and converted to titlecases.
  cleaned_name = s["name"].strip().title()
  #cleaned_roll has rollno convert into integer
  cleaned_roll = int(s["roll"])
  #cleaned_marks has list of marks converted to integer
  cleaned_marks =[int(marks) for marks in s["marks_str"].split(",")]
  #name_validation is used to check whether name contains only alphabets
  name_validation = cleaned_name.replace(" ","").isalpha()
  validation_message = "✓ Valid name" if name_validation else "✗ Invalid name"
  #storing cleaned student details
  student_details ={
      "name":cleaned_name,
      "roll":cleaned_roll,
      "marks":cleaned_marks
  }
  cleaned_students.append(student_details)
  # displaying name validation process
  print(f"{cleaned_name} {validation_message}")
  # displaying formatted student details
  print(f"{"="*30}")
  print(f"Name: {cleaned_name}")
  print(f"Roll: {cleaned_roll}")
  print(f"Marks: {cleaned_marks}")
  print(f"{"="*30}")
```

**Displaying name of roll.no 103 in upper and lower case**

```python
for cs in cleaned_students:
    if cs["roll"]==103:
      print(f"Name of roll.no 103 in upper case: {cs['name'].upper()}")
      print(f"Name of roll.no 103 in lower case: {cs['name'].lower()}")
```

# Task 2 — Marks Analysis Using Loops & Conditionals

```python
student_name = "Ayesha Sharma"
subjects     = ["Math", "Physics", "CS", "English", "Chemistry"]
marks        = [88, 72, 95, 60, 78]
# displaying grades for each subjects based on their using for loop:
for i in range(len(subjects)):
  m=marks[i]
  if m >= 90:grade = "A+"
  elif m >= 80:grade = "A"
  elif m >= 70:grade = "B"
  elif m >= 60:grade = "C"
  else:grade ="F"
  print(f"{subjects[i]}: {m} ({grade})")
```

**Calculating and Displaying Sum, Average, Highest and Lowest scoring subject**

```python
# Calculating Sum, Average, Highest and Lowest scoring subject
total_marks = sum(marks)
average_marks = round(total_marks / len(marks),2)
h_marks = marks.index(max(marks))
l_marks = marks.index(min(marks))

# Displaying the calculated values:
print(f"Total Marks: {total_marks}")
print(f"Average Marks: {average_marks}")
print(f"Highest Scoring Subject: {subjects[h_marks]}({marks[h_marks]})")
print(f"Lowest Scoring Subject: {subjects[l_marks]}({marks[l_marks]})")
```

**Simulating a marks-entry system using while loop that allows adding new subject**

```python
#Adding and display new subject marks using while loop
new_subjects_count = 0

while True:
    sub_input = input("\nEnter subject name (or type 'done' to finish): ").strip()

    if sub_input.lower() == 'done':
        break

    marks_input = input(f"Enter marks for {sub_input}: ")

    try:
        m_val = float(marks_input)
        if 0 <= m_val <= 100:
            subjects.append(sub_input)
            marks.append(m_val)
            new_subjects_count += 1
        else:
            print("Warning: Marks must be between 0 and 100.")
    except ValueError:
        print("Warning: Invalid input. Please enter a numeric value for marks.")

# Final Summary
new_avg = sum(marks) / len(marks)
print("\n" + "-"*30)
print(f"New subjects added: {new_subjects_count}")
print(f"Updated overall average: {new_avg:.2f}")
print("-"*30)
```

# Task 3 — Class Performance Summary

```python
class_data = [
    ("Ayesha Sharma",  [88, 72, 95, 60, 78]),
    ("Rohit Verma",    [55, 68, 49, 72, 61]),
    ("Priya Nair",     [91, 85, 88, 94, 79]),
    ("Karan Mehta",    [40, 55, 38, 62, 50]),
    ("Sneha Pillai",   [75, 80, 70, 68, 85]),
]
#Print the Table Header:
# Print Table Header
print(f"{'Name':<15} | {'Average':<7} | {'Status'}")
print("-" * 40)
stu_avg = []
pass_count = 0
fail_count = 0
for name,scores in class_data:
  #calculating the average marks of students
  avg_marks = round(sum(scores)/len(scores),2)
  stu_avg.append((name,avg_marks))
  #Assigning status
  status = "Pass" if avg_marks >= 60 else "Fail"
  # tracking pass/ fail count
  if status == "Pass":
    pass_count += 1
  else:
    fail_count += 1
  # Printing the formatted row
  print(f"{name:<15} |  {avg_marks:>5.2f}  | {status}")

```

**Calculating and Displaying class Average and Class Topper**

```python
# Calculating the class average
class_avg = sum(avg_marks for name,avg_marks in stu_avg)/len(stu_avg)
#Finding the class topper
topper_name, topper_avg = max(stu_avg, key=lambda x: x[1])
print("-"*40)
print(f"Passed : {pass_count}, Failed : {fail_count}")
print(f"Class Topper : {topper_name} ({topper_avg:.2f})")
print(f"Class Average : {class_avg:.2f}")
```

# Task 4 — String Manipulation Utility

```python
essay = "  python is a versatile language. it supports object oriented, functional, and procedural programming. python is widely used in data science and machine learning.  "
# removing the whitespaces using strip
clean_essay = essay.strip()
#printing striped clean essay
print(f"1.Clean Essay :{clean_essay}")
#printing clean essay in title case
print(f"2.Clean Essay in Title Case :{clean_essay.title()}")
#Count the word python (case-insensitive) -- Note Since the essay statement is already in lower case we could count directly.
py_count = clean_essay.count("python")
print(f"3.Count of 'python' : {py_count}")
# Replacing 'python' with "Python 🐍"
replaced_essay = clean_essay.replace("python","Python 🐍")
print(f"4.Replaced Essay : {replaced_essay}")
# Spliting essay into individual sentences
sentences = clean_essay.split(".")
print(f"5.Splitted Essay : {sentences}")
#Numbering the sentences with punctuatuion check
for i,sentence in enumerate(sentences,1):
  sentence = sentence.strip()
  if not sentence.endswith("."):
    sentence += "."
    print(f"{i}. {sentence}")

```
